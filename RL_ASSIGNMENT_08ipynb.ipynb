```python
import numpy as np

# Transition probabilities: transition_prob[state][action] = [next_state, reward, done]
transition_prob = {
    0: {0: [0, 0, False], 1: [1, 0, False]},
    1: {0: [0, 0, False], 1: [1, 1, True]}
}

# Initialize values arbitrarily
V = np.zeros(2)

# Discount factor
gamma = 0.9

# Value Iteration function
def value_iteration():
    global V  # Declare V as a global variable
    num_iterations = 100

    for i in range(num_iterations):
        new_V = np.zeros(2)
        for state in range(2):
            action_values = []
            for action in transition_prob[state]:
                next_state, reward, done = transition_prob[state][action]
                action_values.append(reward + gamma * V[next_state])
            new_V[state] = max(action_values)
        V = new_V  # Update the global V variable
        print(f"Values after iteration {i + 1}: {V}")

# Call the value iteration function
value_iteration()
```
