```python
import numpy as np
```

```python
np.random.seed(1)
```

```python
x=np.random.randint(0,50,10)
print(x)
```

```python
y=x+np.random.randint(0,10,10)
print(y)
```

```python
from ast import increment_lineno
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
matplotlib.style.use('ggplot')
plt.scatter(x,y)
plt.show()
```

```python
import pandas as pd
df=pd.DataFrame({'a':np.random.randint(0,50,10)})
df['b']=df['a']+np.random.randint(0,10,10)
df['c']=100-df['a']+np.random.randint(0,5,10)
df['d']=np.random.randint(0,50,10)
df.corr()
```

```python
pd.plotting.scatter_matrix(df,figsize=(6,6))
plt.show()
```

```python
plt.matshow(df.corr())
plt.xticks(range(len(df.columns)),df.columns)
plt.yticks(range(len(df.columns)),df.columns)
plt.colorbar()
plt.show()

```
